```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict, Optional 
```

```python
# Create a state

class CricketState(TypedDict, total = False):

    runs: int
    balls: int 
    fours: int
    sixes: int

    strike_rate: float
    runs_in_boundary_percentage: float
    balls_per_boundary: float
    summary: str
```

```python
# Create a function for strike rate

def strike_rate(state: CricketState) -> CricketState:

    runs = state['runs']
    balls = state['balls']

    sr = (runs/balls) * 100
    # state['strike_rate'] = sr

    return {'strike_rate': sr}
```

```python
# Create a function for runs in boundary percent 

def runs_in_boundary_percent(state: CricketState) -> CricketState:

    runs = state['runs']
    fours = state['fours']
    sixes = state['sixes']

    rbp = (((fours*4)+(sixes*6))/runs) * 100
    # state['runs_in_boundary_percentage'] = rbp
    
    return {'runs_in_boundary_percentage': rbp}
```

```python
#  Create a function for balls per boundary

def balls_per_boundary(state: CricketState) -> CricketState:

    balls = state['balls']
    fours = state['fours']
    sixes = state['sixes']

    bpb = balls/(fours+sixes)
    # state['balls_per_boundary'] = bpb

    return {'balls_per_boundary': bpb}
```

```python
# Create a summarizer function

def summarize(state: CricketState) -> CricketState:

    state['summary'] = f''' The Strike Rate of the batsman is : {state['strike_rate']}%. \n
                The Runs in Boundary Percentage of the batsman is : {state['runs_in_boundary_percentage']}%. \n
                The Balls per boundary of the batsman is: {state['balls_per_boundary']}.

                ''' 
    return state
```

```python
#  Create a graph
graph = StateGraph(CricketState)

# Add Nodes
graph.add_node('strike_rate', strike_rate)
graph.add_node('runs_in_boundary_percent', runs_in_boundary_percent)
graph.add_node('balls_per_boundary', balls_per_boundary)
graph.add_node('summarize', summarize)

# Add Edges
graph.add_edge(START, 'strike_rate')
graph.add_edge(START, 'runs_in_boundary_percent')
graph.add_edge(START, 'balls_per_boundary')
graph.add_edge('strike_rate', 'summarize')
graph.add_edge('runs_in_boundary_percent', 'summarize')
graph.add_edge('balls_per_boundary', 'summarize')
graph.add_edge('summarize', END)

# Compile
workflow = graph.compile()
```

```python
# Executing the graph 
initial_state = {
    'runs': 100,
    'balls': 55,
    'fours': 8,
    'sixes': 6
}

final_state = workflow.invoke(initial_state)

print(final_state)
```
